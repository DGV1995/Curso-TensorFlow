# Mejorar la predicción de un modelo lineal con NN

```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
import requests
```

```python
session = tf.Session()
```

```python
seed = 2019
tf.set_random_seed(seed)
np.random.seed(seed)
```

```python
url = "https://raw.githubusercontent.com/DGV1995/Curso-TensorFlow/master/datasets/birthweight.dat"
birth_file = requests.get(url)
birth_data = birth_file.text.split("\r\n")

header = birth_data[0].split("\t")
data = [[float(x) for x in y.split("\t") if len(x) >= 1] for y in birth_data[1:] if len(y) >= 1]
```

```python
header
```

```python
y_vals = np.array([x[0] for x in data])
x_vals = np.array([x[1:8] for x in data])
```

```python
x_vals.shape
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

x_train, x_test, y_train, y_test = train_test_split(x_vals, y_vals, test_size=0.8)

x_train = np.nan_to_num(MinMaxScaler().fit_transform(x_train))
x_test = np.nan_to_num(MinMaxScaler().fit_transform(x_test))
```

```python
batch_size = 80

x_data = tf.placeholder(shape=[None, 7], dtype=tf.float32)
y_target = tf.placeholder(shape=[None, 1], dtype=tf.float32)
```

```python
def init_variables(shape):
    return tf.Variable(tf.random_normal(shape=shape))
```

```python
def logistic(input_layer, weight, bias, activation_sig=True):
    result = tf.add(tf.matmul(input_layer, weight), bias)
    if activation_sig:
        return tf.nn.sigmoid(result)
    else:
        return result
```

Definimos las capas de la red:
- Primera capa ==> Entrada: 7 variables $\rightarrow$ Capa oculta 1: 14 nodos
- Segunda capa ==> Capa oculta 1: 14 nodos $\rightarrow$ Capa oculta 2: 5 nodos
- Última capa ==> Capa oculta 2: 5 nodos $\rightarrow$ Capa de salida: 1 variable

```python
A1 = init_variables(shape=[7,14])
b1 = init_variables(shape=[14])
layer1 = logistic(x_data, A1, b1, True)

A2 = init_variables(shape=[14,5])
b2 = init_variables(shape=[5])
layer2 = logistic(layer1, A2, b2, True)

A3 = init_variables(shape=[5,1])
b3 = init_variables(shape=[1])
final_output = logistic(layer2, A3, b3, False)
```

```python
loss = tf.reduce_mean(tf.nn.sigmoid_cross_entropy_with_logits(logits=final_output, labels=y_target))
optimizer = tf.train.AdamOptimizer(0.002)
train = optimizer.minimize(loss)

init = tf.global_variables_initializer()
session.run(init)
```

```python
prediction = tf.round(tf.nn.sigmoid(final_output)) # 0.0 ó 1.0
accuracy = tf.reduce_mean(tf.cast(tf.equal(prediction, y_target), tf.float32))
```

```python
train_loss_vector = []
test_loss_vector = []

train_acc = []
test_acc = []

for i in range(300):
    rand_idx = np.random.choice(len(x_train), size=batch_size)
    rand_x = x_train[rand_idx]
    rand_y = np.transpose([y_train[rand_idx]])
    session.run(train, feed_dict={x_data: rand_x, y_target: rand_y})
    
    temp_train_loss = session.run(loss, feed_dict={x_data: rand_x, y_target: rand_y})
    temp_test_loss = session.run(loss, feed_dict={x_data: x_test, y_target: np.transpose([y_test])})
    train_loss_vector.append(temp_train_loss)
    test_loss_vector.append(temp_test_loss)
    
    temp_train_acc = session.run(accuracy, feed_dict={x_data: rand_x, y_target: rand_y})
    temp_test_acc = session.run(accuracy, feed_dict={x_data: x_test, y_target: np.transpose([y_test])})
    train_acc.append(temp_train_acc)
    test_acc.append(temp_test_acc)
    
    if (i+1)%25 == 0:
        print("Step #{} ==> Loss: {}, Accuracy: {}".format(i+1, temp_train_loss, temp_train_acc))
```

```python
plt.figure(figsize=(15,8))
plt.plot(train_loss_vector, "b", label="Entrenamiento")
plt.plot(test_loss_vector, "r", label="Test")
plt.title("Función de pérdidas", fontsize=15)
plt.xlabel("Iteración", fontsize=14)
plt.ylabel("Pérdidas", fontsize=14)
plt.legend(loc="upper left")
plt.show()
```

```python
plt.figure(figsize=(15,8))
plt.plot(train_acc, "r-", label="Entrenamiento")
plt.plot(test_acc, "g--", label="Test")
plt.title("Función de precisión", fontsize=15)
plt.xlabel("Iteración", fontsize=14)
plt.ylabel("Pérdidas", fontsize=14)
plt.legend(loc="upper left")
plt.show()
```
