# SVM para problemas de regresión

$$ Loss = \frac{1}{n} \sum_{i=1}^n\max(0, |y_i - y_{pred}| - \epsilon)$$
- $ y_{pred} = Ax + b $


- $ \epsilon $: la mitad del margen, que fuerza a que la función de pérdidas sea 0 para todos los puntos dentro del corredor

```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
import requests
from sklearn import datasets
```

```python
session = tf.Session()
```

```python
iris = datasets.load_iris()
iris
```

```python
x_vals = np.array([x[3] for x in iris.data])
y_vals = np.array([y[0] for y in iris.data])
```

```python
plt.figure(figsize=(10,6))
plt.plot(x_vals, y_vals, "bo")
plt.show()
```

```python
train_idx = np.random.choice(len(x_vals), round(len(x_vals) * 0.8), replace=False)
test_idx = np.array(list(set(range(len(x_vals))) - set(train_idx)))

x_train = x_vals[train_idx]
y_train = y_vals[train_idx]

x_test = x_vals[test_idx]
y_test = y_vals[test_idx]
```

```python
batch_size = 50

x_data = tf.placeholder(shape=[None, 1], dtype=tf.float32)
y_target = tf.placeholder(shape=[None, 1], dtype=tf.float32)

A = tf.Variable(tf.random_normal(shape=[1,1]))
b = tf.Variable(tf.random_normal(shape=[1,1]))

y_pred = tf.add(tf.matmul(x_data, A), b)
```

```python
epsilon = tf.constant([0.5]) # Anchura del corredor
loss = tf.reduce_mean(tf.maximum(0.0, tf.subtract(tf.abs(tf.subtract(y_target, y_pred)), epsilon)))
```

```python
optimizer = tf.train.GradientDescentOptimizer(learning_rate = 0.05)
train = optimizer.minimize(loss)
```

```python
init = tf.global_variables_initializer()
session.run(init)
```

```python
train_loss_vector = []
test_loss_vector = []

for i in range(300):
    rand_idx = np.random.choice(len(x_train), size=batch_size)
    rand_x = np.transpose([x_train[rand_idx]])
    rand_y = np.transpose([y_train[rand_idx]])
    session.run(train, feed_dict={x_data: rand_x, y_target: rand_y})
    
    train_temp_loss = session.run(loss, feed_dict={x_data: np.transpose([x_train]), 
                                                   y_target: np.transpose([y_train])})
    train_loss_vector.append(train_temp_loss)
    
    test_temp_loss = session.run(loss, feed_dict={x_data: np.transpose([x_test]), 
                                                  y_target: np.transpose([y_test])})
    test_loss_vector.append(test_temp_loss)
    
    if (i+1)%50 == 0:
        print("Step #{} ==> A = {}, b = {}".format(i+1, session.run(A), session.run(b)))
        print("Loss Train = {}, Loss Test = {}". format(train_temp_loss, test_temp_loss))
```

```python
[[slope]] = session.run(A)
[[intercept]] = session.run(b)
[width] = session.run(epsilon)

print("y = " + str(np.round(slope, 4)) + "x " + str(np.round(intercept, 4)))
```

```python
plt.figure(figsize=(14,9))
plt.plot(train_loss_vector, "k-", label="Entrenamiento")
plt.plot(test_loss_vector, "r--", label="Testing")
plt.title("Función de pérdidas en entrenamiento y testing", fontsize=15)
plt.xlabel("Iteración", fontsize=13)
plt.ylabel("Pérdidas", fontsize=13)
plt.legend(loc="upper right")
plt.show()
```

```python
best_fit = []
best_fit_upper = []
best_fit_lower = []

for x in x_vals:
    best_fit.append(slope*x + intercept)
    best_fit_upper.append(slope*x + intercept + width)
    best_fit_lower.append(slope*x + intercept - width)
```

```python
plt.figure(figsize=(14,9))
plt.plot(x_vals, y_vals, "bo", label="Datos")
plt.plot(x_vals, best_fit, "r-", label="SVM Regresión lineal", linewidth=3)
plt.plot(x_vals, best_fit_upper, "r--", linewidth=0.5)
plt.plot(x_vals, best_fit_lower, "r--", linewidth=0.5)
plt.title("Predicción usando SVM", fontsize=15)
plt.xlabel("Anchura de pétalo", fontsize=13)
plt.ylabel("Longitud de sépalo", fontsize=13)
plt.ylim([0,10])
plt.legend(loc="upper left")
plt.show()
```
