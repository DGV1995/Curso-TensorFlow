# Predicción de la longitud del sépalo en función de la anchura del pétalo

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
from sklearn import datasets
```

```python
session = tf.Session()
```

```python
iris = datasets.load_iris()
```

```python
x_vals = np.array([x[3] for x in iris.data]) # Longitud del péatlo
y_vals = np.array([x[0] for x in iris.data]) # Anchura del sépalo
```

```python
plt.plot(x_vals, y_vals, "o")
plt.show()
```

```python
learning_rate = 0.05
batch_size = 25
x_data = tf.placeholder(shape=[None, 1], dtype=tf.float32)
y_target = tf.placeholder(shape=[None, 1], dtype=tf.float32)
A = tf.Variable(tf.random_normal(shape=[1,1]))
b = tf.Variable(tf.random_normal(shape=[1,1]))
```

$$ y = Ax + b $$

```python
y_pred = tf.add(tf.matmul(x_data, A), b)
loss = tf.reduce_mean(tf.square(y_pred - y_target))
optimizer = tf.train.GradientDescentOptimizer(learning_rate=learning_rate)
train = optimizer.minimize(loss)
init = tf.global_variables_initializer()
session.run(init)
```

```python
loss_vect = []

for i in range(100):
    rand_idx = np.random.choice(len(x_vals), size=batch_size)
    rand_x = np.transpose([x_vals[rand_idx]])
    rand_y = np.transpose([y_vals[rand_idx]])
    session.run(train, feed_dict={x_data: rand_x, y_target: rand_y})
    temp_loss = session.run(loss, feed_dict={x_data: rand_x, y_target: rand_y})
    loss_vect.append(temp_loss)
    if (i+1)%10 == 0:
        print("Step #{} ==> A = {}, b = {}, Loss = {}".format(i+1, session.run(A), session.run(b), temp_loss))
```

```python
[[slope]] = session.run(A)
[[intercept]] = session.run(b)
```

```python
print(str(np.round(slope, 3)) + "x + " + str(np.round(intercept,3)))
```

```python
best_fit = []

for x in x_vals:
    best_fit.append(slope*x + intercept)
```

```python
plt.figure(figsize=(12,7))
plt.plot(x_vals, y_vals, "o", label="Datos")
plt.plot(x_vals, best_fit, "r-", label="Recta de regresión con TensorFlow", linewidth=3)
plt.xlabel("Anchura del pétalo", fontsize=14)
plt.ylabel("Longitud del sépalo", fontsize=14)
plt.title("Longitud del sépalo vs anchura del pétalo", fontsize=15)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(loss_vect, "k-")
plt.title("Función de pérdidas con L2")
plt.xlabel("Iteración del algoritmo")
plt.ylabel("Medida de L2")
plt.show()
```
