# Funciones de pérdida
Una función de pérdida indica a tensorflow cómo de buena o de mala es una predicción, en comparación con el resultado que uno desea

```python
import tensorflow as tf
session = tf.Session()
```

```python
import matplotlib.pyplot as plt
```

```python
x_vals = tf.linspace(-1.,1.,500) # Predicción
target = tf.constant(0.) # Valores reales
```

# Tipos de funciones de pérdidas en problemas de predicción (regresión)

## Norma en L2 (distancia Euclídea)
* Continua
* Derivable

$$ L2(y_{real},{y_pred}) = (y_{real} - y_{pred})^2 $$

```python
l2_y_vals = tf.square(target - x_vals)
```

```python
l2_y_ouput = session.run(l2_y_vals)
print(l2_y_ouput)
```

```python
print(session.run(tf.nn.l2_loss(target - x_vals)))
```

## Norma en L1 (valor absoluto)
* Continua
* No derivable

$$ L1(y_{real}, y_{pred}) = |y_{real} - y_{pred}| $$

```python
l1_y_vals = tf.abs(target - x_vals)
l1_y_ouput = session.run(l1_y_vals)
print(l1_y_ouput)
```

## Pseudo-Huber
* Continua
* Derivable

$$ PH(y_{real}, y_{pred}) = \delta^2 \cdot \sqrt{1 + \left(\frac{y_{real}-y_{pred}}{\delta}\right)^2}-1,  \delta > 0 $$

```python
delta1 = tf.constant(0.25)
phuber1_y_vals = tf.multiply(tf.square(delta1), tf.sqrt(1. + tf.square((target-x_vals)/delta1))-1.)
phuber1_y_output = session.run(phuber1_y_vals)
print(phuber1_y_output)
```

```python
delta2 = tf.constant(5.0)
phuber2_y_vals = tf.multiply(tf.square(delta2), tf.sqrt(1. + tf.square((target-x_vals)/delta2))-1.)
phuber2_y_output = session.run(phuber2_y_vals)
print(phuber2_y_output)
```

# Tipos de funciones de pérdidas para problemas de clasificación

```python
x_vals = tf.linspace(-3., 5., 500)
target = tf.constant(1.)
targets = tf.fill([500, ], 1.)
```

```python
print(session.run(targets))
```

## Hinge (función visagra)

$$ H(y_{real}, y_{pred}) = max(0, 1 - y_{real}\cdot y_{pred}) $$

```python
hinge_y_vals = tf.maximum(0., 1. - tf.multiply(target, x_vals))
hinge_y_out = session.run(hinge_y_vals)
print(hinge_y_out)
```

## Entropía cruzada (función logística)

$$ H(y_{real}, y_{pred}) = -y_{real} \cdot log(y_{pred}) - (1-y_{real}) \cdot log(1 - y_{pred}) $$

```python
xentr_y_vals = -tf.multiply(target, tf.log(x_vals)) - tf.multiply(1.-target, tf.log(1.-x_vals))
xentr_y_out = session.run(xentr_y_vals)
print(xentr_y_out)
```

## Entropía cruzada de Sigmoide

$$ H(y_{real}, y_{pred}) = -y_{real} \cdot log\left(\frac{1}{1+e^{-y_{pred}}}\right) - (1-y_{real}) \cdot log\left(1 - \frac{1}{1+e^{-y_{pred}}}\right) $$

```python
xentr_sig_y_vals = tf.nn.sigmoid_cross_entropy_with_logits(logits=x_vals, labels=targets)
xentr_sig_y_out = session.run(xentr_sig_y_vals)
print(xentr_sig_y_out)
```

```python
pos_weight = tf.constant(0.5) # Possitive weight
xentr_sig_w_y_vals = tf.nn.weighted_cross_entropy_with_logits(logits=x_vals, targets=targets, pos_weight=pos_weight)
xentr_sig_w_y_out = session.run(xentr_sig_w_y_vals)
print(xentr_sig_w_y_out)
```

## Softmax Cross Entropy
Cuando solo hay una variable categórica a predecir en lugar de múltiples de ellas

$$ SM(z_k) = \frac{e^{z_k}}{\sum_{i=1}^n e^{z_i}} $$

```python
unscaled_logits = tf.constant([[1.,-3.,10.]])
target_dist = tf.constant([[0.1,0.02,0.88]])

softmax_xentr = tf.nn.softmax_cross_entropy_with_logits_v2(logits=unscaled_logits, labels=target_dist)
print(session.run(softmax_xentr))
```

## Sparse softmax cross entropy

```python
unscaled_logits = tf.constant([[1.,-3.,10.]])
sparsed_target_dist = tf.constant([2])

sparsed_xentr = tf.nn.sparse_softmax_cross_entropy_with_logits(logits=unscaled_logits,
                                                                   labels=sparsed_target_dist)
print(session.run(sparsed_xentr))
```

# Evaluar las funciones de pérdidas
### Función de pérdida para regresión

```python
x_vals = tf.linspace(-1.,1.,500)
x_array = session.run(x_vals)

plt.plot(x_array, l2_y_ouput, "r-", label="Función L2")
plt.plot(x_array, l1_y_ouput, "g--", label="Función L1")
plt.plot(x_array, phuber1_y_output, "b-.", label="Pseudo Huber (0.25)")
plt.plot(x_array, phuber2_y_output, "k:", label="Pseudo Huber (5.0)")

plt.ylim(-0.4, 0.5)
plt.legend(loc="lower right", prop={"size":10})

plt.show()
```

### Función de pérdida para clasificación

```python
x_vals = tf.linspace(-3., 5., 500)
x_array = session.run(x_vals)

plt.plot(x_array, hinge_y_out, "r-", label="Función Hinge")
plt.plot(x_array, xentr_y_out, "g--", label="Cross Entropy")
plt.plot(x_array, xentr_sig_y_out, "b-.", label="Cross Entropy Sigmoid")
plt.plot(x_array, xentr_sig_w_y_out, "k:", label="Weighted Cross Entropy (w=0.5)")

plt.ylim(-2, 3)
plt.legend(loc="lower right", prop={"size":10})

plt.show()
```
